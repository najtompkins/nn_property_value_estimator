```python
from redfin import Redfin
import requests
import os
from pprint import pprint
```

```python
try:
    client = Redfin()

    address = '110 N Federal Hwy #616'

    response = client.search(address)
    url = response['payload']['exactMatch']['url']
    initial_info = client.initial_info(url)

    property_id = initial_info['payload']['propertyId']
    mls_data = client.below_the_fold(property_id)

    listing_id = initial_info['payload']['listingId']
    avm_details = client.avm_details(property_id, listing_id)

    pprint(address)
    pprint(f'Response: {response}')
    pprint(f'url: {url}')
    pprint(f'initial info: {initial_info}')
    pprint(f'property id: {property_id}')
    pprint(f'mls data: {mls_data}')
    pprint(f'listing id: {listing_id}')
    pprint(f'avm details: {avm_details}')
except Exception as e:
    pprint(f'error: {e}')
```

```python
# Step 1: Search for the address
address = '110 N Federal Hwy #616'
client = Redfin()
search_response = client.search(address)
url = search_response['payload']['exactMatch']['url']
```

```python
# Step 2: Extract propertyId and listingId
initial_info = client.initial_info(url)
property_id = initial_info['payload']['propertyId']
listing_id = initial_info['payload']['listingId']
```

```python
pprint(avm_details)
```

```python

# Step 3: Get property details including photo URLs
avm_details = client.avm_details(property_id, listing_id)
photo_urls = avm_details['payload']['photos']
```

```python

# Step 4: Download the photos
download_folder = 'redfin_photos'
os.makedirs(download_folder, exist_ok=True)
```

```python

for i, photo_url in enumerate(photo_urls, start=1):
    response = requests.get(photo_url)
    if response.status_code == 200:
        with open(os.path.join(download_folder, f'photo_{i}.jpg'), 'wb') as f:
            f.write(response.content)
        print(f'Downloaded photo_{i}.jpg')
    else:
        print(f'Failed to download photo_{i}.jpg')

print('Photo download complete!')
```
